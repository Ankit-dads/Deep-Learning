```python
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
df = pd.read_csv("Sentiment Analysis Dataset.csv",encoding='latin1')
df.head(10)
```

```python
sentence = df['SentimentText']
labels = df['Sentiment']
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentence)
sequences = tokenizer.texts_to_sequences(sentence)
print(sequences[:5])
```

```python
X = pad_sequences(sequences, maxlen=5)

print(X)
```

```python
y = np.array(labels)
y
```

```python
model = Sequential([
    Embedding(input_dim=50, output_dim=8, input_length=5),

    SimpleRNN(16),

    Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
    X,
    y,
    epochs=30,
    verbose=1,
    batch_size = 4096
)
```

```python
test = ["I love this film"]

seq = tokenizer.texts_to_sequences(test)

pad = pad_sequences(seq, maxlen=50)

prediction = model.predict(pad)

print(prediction)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(X, y, verbose=0)

print("Loss:", loss)
print("Accuracy:", accuracy)
print("Accuracy (%):", accuracy * 100)
```
