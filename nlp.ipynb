```python
df = pd.read_csv('ham_spam_nlp_dataset_1000.csv')
df.head()
```

```python
df['Email'] = df['email'].str.lower()
df.drop(['email'],axis = 1)
```

```python
df.head()
```

```python
import string
```

```python
def remove_punct(email):
    clean = [c for c in email if c not in string.punctuation]
    return"".join(clean)

df['clean_msg'] = df['email'].apply(remove_punct)
```

```python
df.head()
```

```python
nltk.download("punkt")
from nltk.tokenize import word_tokenize 



df["tokens"] = df["clean_msg"].apply(word_tokenize)
print(df["tokens"].head(2).values)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
def filter_stops(token_list):
    return[w for w in token_list if w not in stop_words]
df["clean_tokens"] =  df["tokens"].apply(filter_stops)
print(df['clean_tokens'].head(5).values)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
def lemmatize_tokens(token_list):
    return[lemmatizer.lemmatize(w) for w in token_list]

df['final_tokens'] = df['clean_tokens'].apply(lemmatize_tokens)
print(df['final_tokens'].head(5).values)
```
