```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('tested.csv')
```

```python
df.head()
```

```python
df.drop(columns=['Pclass','PassengerId', 'Name', 'Ticket', 'Cabin','SibSp','Ticket','Parch','Sex','Embarked'],inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
df.isnull().sum()
```

```python
x = df[['Age','Fare']]
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=24)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
y_pred = model.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test,y_pred)
confusion_matrix = confusion_matrix(y_test,y_pred)
print(accuracy)
print(confusion_matrix)
```

```python
single_n = Sequential(
    [Dense(1, input_dim = 2, activation = 'sigmoid')])
```

```python
single_n.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy'])
```

```python
history = single_n.fit(
    x_train_scaled,
    y_train,
    epochs = 500,
    batch_size = 64,
    verbose = 1
)
```

```python
n_predict = single_n.predict(x_test_scaled)
```

```python
n_predict = single_n.predict(x_test_scaled)

# Convert probabilities → 0 or 1
n_predict_binary = (n_predict > 0.5).astype(int).flatten()

print(precision_score(y_test, n_predict_binary))
```
